```python
import pandas as pd
df = pd.read_csv("/kaggle/input/titanic-train-dataset/train.csv")
df.head()
```

1. Kazada ölenlerin yaş ortalaması

```python
df[df["Survived"] == 0]["Age"].mean()
```

 2. Kazada ölenlerin bilet fiyatlarının ortalaması ve medyanı

```python
fare_dead = df[df["Survived"] == 0]["Fare"]
print("Ortalama:", fare_dead.mean())
print("Medyan:", fare_dead.median())
```

  3.Kazada ölen erkeklerin yaş ortalaması

```python
df[(df["Survived"] == 0) & (df["Sex"] == "male")]["Age"].mean()
```

 4. Kazada ölen kadınların yaş ortalaması

```python
df[(df["Survived"] == 0) & (df["Sex"] == "female")]["Age"].mean()
```

5. Kazadan kurtulanların yaş ortalaması

```python
df[df["Survived"] == 1]["Age"].mean()
```

6. Kazadan kurtulanların bilet fiyatlarının ortalaması

```python
df[df["Survived"] == 1]["Fare"].mean()
```

7. Kazadan kurtulan toplam kişi sayısı

```python
df[df["Survived"] == 1].shape[0]
# veya:
len(df[df["Survived"] == 1])
```

 8. 10 yaşından küçüklerin bilet fiyatlarının medyan değeri

```python
df[df["Age"] < 10]["Fare"].median()
```

 9. 1., 2. ve 3. sınıf bilet fiyatlarının ortalama ve medyanlarını karşılaştırma

```python
df.groupby("Pclass")["Fare"].agg(["mean", "median"])
```

10. Kazada ölen kadınların oranı ile erkeklerin oranını karşılaştırma

```python
dead_female_ratio = len(df[(df["Sex"] == "female") & (df["Survived"] == 0)]) / len(df[df["Sex"] == "female"])
dead_male_ratio = len(df[(df["Sex"] == "male") & (df["Survived"] == 0)]) / len(df[df["Sex"] == "male"])

print("Kadınların ölüm oranı:", dead_female_ratio)
print("Erkeklerin ölüm oranı:", dead_male_ratio)
```
